```python
import pandas as pd
import json
import glob

df_conjunto = pd.DataFrame()
 
# Opening JSON file
for file in files_tot:
    f = open(file, encoding='utf-8')

    data = json.load(f)

    url = []
    nombre = []
    ciudad = []
    precio = []
    latitud = []
    longitud = []
    rating = []
    reviews = []
    cerrado = []
    cerrado_temporal = []
    horarios = []
    ocupacion = []
    lgbt = []
    aperitivos = []
    terraza = []
    cerveza = []
    desayuno_almuerzo = []
    sentarse = []
    llevar = []
    postres = []
    acepta_reserva = []
    perros = []
    perros_fuera = []
    wifi = []
    wifi_gratis = []
    vino = []

    for i, cafeteria in enumerate(data):
        url.append(cafeteria.get('url'))
        nombre.append(cafeteria.get('title'))
        ciudad.append(cafeteria.get('city'))
        precio.append(cafeteria.get('price'))
        latitud.append(cafeteria.get('location').get('lat'))
        longitud.append(cafeteria.get('location').get('lng'))
        rating.append(cafeteria.get('totalScore'))
        reviews.append(cafeteria.get('reviewsCount'))
        cerrado.append(cafeteria.get('permanentlyClosed'))
        cerrado_temporal.append(cafeteria.get('temporarilyClosed'))      
        horarios.append(cafeteria.get('openingHours'))     
        ocupacion.append(cafeteria.get('popularTimesHistogram'))

        if 'Público usual' in cafeteria.get('additionalInfo', {}):
            for item in cafeteria['additionalInfo']['Público usual']:
                if 'Amigable con LGBTQ+' in item and item['Amigable con LGBTQ+']:
                    lgbt.append(item['Amigable con LGBTQ+'])

                    break  # Terminar el bucle si se encuentra la clave
            else:
                lgbt.append(False)
        else:
            lgbt.append(False)


        if 'Qué ofrece' in cafeteria.get('additionalInfo', {}):
            for item in cafeteria['additionalInfo']['Qué ofrece']:
                if 'Aperitivos' in item and item['Aperitivos']:
                    aperitivos.append(item['Aperitivos'])

                    break  # Terminar el bucle si se encuentra la clave
            else:
                aperitivos.append(False)
        else:
            aperitivos.append(False)

        try:
            if 'Opciones de servicio' in cafeteria.get('additionalInfo', {}):
                for item in cafeteria['additionalInfo']['Opciones de servicio']:
                    if 'Asientos al aire libre' in item and item['Asientos al aire libre']:
                        terraza.append(item['Asientos al aire libre'])

                        break  # Terminar el bucle si se encuentra la clave
                else:
                    terraza.append(False)
            else:
                terraza.append(False)
        except:
            terraza.append(False)

        if 'Qué ofrece' in cafeteria.get('additionalInfo', {}):
            for item in cafeteria['additionalInfo']['Qué ofrece']:
                if 'Cerveza' in item and item['Cerveza']:
                    cerveza.append(item['Cerveza'])

                    break  # Terminar el bucle si se encuentra la clave
            else:
                cerveza.append(False)
        else:
            cerveza.append(False)

        if 'Opciones del local' in cafeteria.get('additionalInfo', {}):
            for item in cafeteria['additionalInfo']['Opciones del local']:
                if ('Desayunos' in item and item['Desayunos']) or ('Desayuno-almuerzo' in item and item['Desayuno-almuerzo']) or ('Almuerzo' in item and item['Almuerzo']):
                    desayuno_almuerzo.append(True)

                    break  # Terminar el bucle si se encuentra la clave
            else:
                desayuno_almuerzo.append(False)
        else:
            desayuno_almuerzo.append(False)

        # Espacio para sentarse está mal indexado en la info, no refleja bien la realidad. Eliminable. 
        if 'Opciones del local' in cafeteria.get('additionalInfo', {}):
            for item in cafeteria['additionalInfo']['Opciones del local']:
                if 'Espacio con asientos' in item and item['Espacio con asientos']:
                    sentarse.append(item['Espacio con asientos'])

                    break  # Terminar el bucle si se encuentra la clave
            else:
                sentarse.append(False)
        else:
            sentarse.append(False)

        try:
            # Probablemente no indica que el café sea para llevar, si no si tienen comida para llevar. Eliminable
            if 'Opciones de servicio' in cafeteria.get('additionalInfo', {}):
                for item in cafeteria['additionalInfo']['Opciones de servicio']:
                    if 'Para llevar' in item and item['Para llevar']:
                        llevar.append(item['Para llevar'])

                        break  # Terminar el bucle si se encuentra la clave
                else:
                    llevar.append(False)
            else:
                llevar.append(False)
        except:
            llevar.append(False)

        if 'Opciones del local' in cafeteria.get('additionalInfo', {}):
            for item in cafeteria['additionalInfo']['Opciones del local']:
                if 'Postres' in item and item['Postres']:
                    postres.append(item['Postres'])

                    break  # Terminar el bucle si se encuentra la clave
            else:
                postres.append(False)
        else:
            postres.append(False)

        if 'Planificación' in cafeteria.get('additionalInfo', {}):
            for item in cafeteria['additionalInfo']['Planificación']:
                if 'Se aceptan reservas' in item and item['Se aceptan reservas']:
                    acepta_reserva.append(item['Se aceptan reservas'])

                    break  # Terminar el bucle si se encuentra la clave
            else:
                acepta_reserva.append(False)
        else:
            acepta_reserva.append(False)

        if 'Servicios' in cafeteria.get('additionalInfo', {}):
            for item in cafeteria['additionalInfo']['Servicios']:
                if 'Se permiten perros' in item and item['Se permiten perros']:
                    perros.append(item['Se permiten perros'])

                    break  # Terminar el bucle si se encuentra la clave
            else:
                perros.append(False)
        else:
            perros.append(False)

        if 'Mascotas' in cafeteria.get('additionalInfo', {}):
            for item in cafeteria['additionalInfo']['Mascotas']:
                if 'Se permiten perros afuera' in item and item['Se permiten perros afuera']:
                    perros_fuera.append(item['Se permiten perros afuera'])

                    break  # Terminar el bucle si se encuentra la clave
            else:
                perros_fuera.append(False)
        else:
            perros_fuera.append(False)

        if 'Servicios' in cafeteria.get('additionalInfo', {}):
            for item in cafeteria['additionalInfo']['Servicios']:
                if 'Wi-Fi' in item and item['Wi-Fi']:
                    wifi.append(item['Wi-Fi'])

                    break  # Terminar el bucle si se encuentra la clave
            else:
                wifi.append(False)
        else:
            wifi.append(False)

        if 'Servicios' in cafeteria.get('additionalInfo', {}):
            for item in cafeteria['additionalInfo']['Servicios']:
                if 'Tiene Wi-Fi gratis' in item and item['Tiene Wi-Fi gratis']:
                    wifi_gratis.append(item['Tiene Wi-Fi gratis'])

                    break  # Terminar el bucle si se encuentra la clave
            else:
                wifi_gratis.append(False)
        else:
            wifi_gratis.append(False)

        if 'Qué ofrece' in cafeteria.get('additionalInfo', {}):
            for item in cafeteria['additionalInfo']['Qué ofrece']:
                if 'Vino' in item and item['Vino']:
                    vino.append(item['Vino'])

                    break  # Terminar el bucle si se encuentra la clave
            else:
                vino.append(False)
        else:
            vino.append(False)

    # Construímos DF
    datos = {
        'url': url,
        'nombre': nombre,
        'ciudad': ciudad,
        'precio': precio,
        'latitud': latitud,
        'longitud': longitud,
        'rating': rating,
        'reviews': reviews,
        'cerrado': cerrado,
        'cerrado_temporal': cerrado_temporal,
        'horarios': horarios,
        'ocupacion': ocupacion,
        'lgbt': lgbt,
        'aperitivos': aperitivos,
        'terraza': terraza,
        'cerveza': cerveza,
        'desayuno_almuerzo': desayuno_almuerzo,
        'sentarse': sentarse,
        'llevar': llevar,
        'postres': postres,
        'acepta_reserva': acepta_reserva,
        'perros': perros,
        'perros_fuera': perros_fuera,
        'wifi': wifi,
        'wifi_gratis': wifi_gratis,
        'vino': vino
    }

    # Convertir el diccionario en un DataFrame de pandas
    df_temporal = pd.DataFrame(datos)
    
    df_conjunto = pd.concat([df_conjunto, df_temporal], ignore_index=True)
    
df_conjunto
```
